```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, SpatialDropout1D
from sklearn.preprocessing import LabelEncoder

# 1. Load Data (eb2a et2kd mn masar el files bta3tk)
train_df = pd.read_csv(r"C:\Users\paula\OneDrive\Documentos\GitHub\Smart-Waste-Classification-NN-PROJECT-\processed\train.csv")
val_df = pd.read_csv(r"C:\Users\paula\OneDrive\Documentos\GitHub\Smart-Waste-Classification-NN-PROJECT-\processed\val.csv")
test_df = pd.read_csv(r"C:\Users\paula\OneDrive\Documentos\GitHub\Smart-Waste-Classification-NN-PROJECT-\processed\test.csv")

# Bneftred en esm el column bta3 el klam esmo 'text' w el class esmo 'label'
# Lw asameehom mkhtalefa 3andk, 8ayarhom hena

X_train_text = train_df['text_description'].astype(str)
y_train_label = train_df['label']

X_val_text = val_df['text_description'].astype(str)
y_val_label = val_df['label']

X_test_text = test_df['text_description'].astype(str)
y_test_label = test_df['label']

# 2. Encode Labels (N7wl el classes zay 'glass', 'metal' lـ arqam 0, 1, 2, 3)
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train_label)
y_val = label_encoder.transform(y_val_label)
y_test = label_encoder.transform(y_test_label)

# 3. Tokenization & Padding (N7wl el text lـ arqam 3ashan el model yfhmha)
MAX_WORDS = 5000 # Akbr 3adad mn el kalimat el mohema
MAX_LEN = 100    # Tol el gomla el wa7da (lw atwl ht-t2s, lw aqal hyzed-ha asfar)

tokenizer = Tokenizer(num_words=MAX_WORDS, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train_text)

X_train = pad_sequences(tokenizer.texts_to_sequences(X_train_text), maxlen=MAX_LEN)
X_val = pad_sequences(tokenizer.texts_to_sequences(X_val_text), maxlen=MAX_LEN)
X_test = pad_sequences(tokenizer.texts_to_sequences(X_test_text), maxlen=MAX_LEN)

# 4. Build LSTM Model (Hena bnbny el architecture bta3 el model)
num_classes = len(np.unique(y_train))

model = Sequential()
model.add(Embedding(input_dim=MAX_WORDS, output_dim=128, input_length=MAX_LEN))
model.add(SpatialDropout1D(0.2)) # 3ashan n2ll el Overfitting
model.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2)) # El LSTM layer
model.add(Dense(num_classes, activation='softmax')) # Output layer

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()

# 5. Train Model (Hnkhly el model yzakr mn el train.csv w ymthn nfso fe el val.csv)
epochs = 10
batch_size = 32

history = model.fit(
    X_train, y_train, 
    epochs=epochs, 
    batch_size=batch_size, 
    validation_data=(X_val, y_val)
)

# 6. Test Model (Hena bnt2kd mn mostwa el model 3la el test.csv elli hoa mshafosh abl keda)
print("\n--- Testing the Model on test.csv ---")
loss, accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {accuracy*100:.2f}%")
```

```python
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add
from tensorflow.keras.models import Model

# --- 1. Load Data w N7azhz el Text ---
train_df = pd.read_csv(r"processed/train.csv")

# Bn7ot 'startseq' fe awl el gomla w 'endseq' fe akherha
# 3ashan el LSTM y3rf emta ybd2 yktb w emta yo2af
train_df['caption'] = 'startseq ' + train_df['text_description'].astype(str) + ' endseq'
captions = train_df['caption'].tolist()

# --- 2. Tokenization (N3ml Qamos el Kalimat) ---
tokenizer = Tokenizer()
tokenizer.fit_on_texts(captions)

# 3adad el kalimat el mokhtalefa fel data bta3tk (+1 3ashan el padding padding byakhod raqam 0)
vocab_size = len(tokenizer.word_index) + 1 

# N7sb atwl description (3adad el kalimat) 3ashan n7dd 7agm el Input
max_length = max(len(c.split()) for c in captions)

print(f"Vocab Size (3adad el kalimat): {vocab_size}")
print(f"Max Length (Atwl gomla): {max_length}")

# --- 3. Build Model (CNN + LSTM) ---
# El Goz2 bta3 el Sora (Image Features mn ResNet50 masalan byb2a 7agmo 2048)
inputs1 = Input(shape=(2048,), name="Image_Features")
fe1 = Dropout(0.5)(inputs1)
fe2 = Dense(256, activation='relu')(fe1)

# El Goz2 bta3 el Text (Sequence Processor - LSTM)
inputs2 = Input(shape=(max_length,), name="Text_Sequence")
se1 = Embedding(vocab_size, 256, mask_zero=True)(inputs2)
se2 = Dropout(0.5)(se1)
se3 = LSTM(256)(se2) # El LSTM

# Dmg el Sora m3 el Text (Decoder)
decoder1 = add([fe2, se3])
decoder2 = Dense(256, activation='relu')(decoder1)
outputs = Dense(vocab_size, activation='softmax', name="Output_Word")(decoder2)

# El Model el Neha2y
model = Model(inputs=[inputs1, inputs2], outputs=outputs)
model.compile(loss='categorical_crossentropy', optimizer='adam')

print("\n--- Model is ready! ---")
model.summary()
```
